```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot  as plt
# from keras.datasets import mnist
from sklearn.decomposition import NMF
```

```python
with np.load('mnist.npz') as data:
    X_train, y_train = data['x_train'], data['y_train']
    X_test, y_test = data['x_test'], data['y_test']

plt.imshow(X_train[0])  
plt.show()
```

```python
model_n = NMF(n_components=8, random_state=42)  
features_n = model_n.fit_transform(X_train[0])  
print("Transformed Data Shape: {}".format(features_n.shape))  
print("Model's Components Shape: {}".format(model_n.components_.shape))
```

```python
def show_as_image(sample):  
	plt.figure()  
	plt.imshow(sample, interpolation='nearest')  
	plt.colorbar()  
	plt.show()  
  
for component in model_n.components_:  
	component = component.reshape(4, 7)  
	show_as_image(component)
```

### Repeat on bigger set

```python
X_train_lim = X_train[:100]  
print("Shape of 100 images train dataset: {}".format(X_train_lim.shape))  
  
X_train_lim = X_train_lim.reshape((100, 28*28))  
print("Shape of 100 images train dataset: {}".format(X_train_lim.shape))
```

```python
model = NMF(n_components=8, random_state=42)  
features = model.fit_transform(X_train_lim)  
  
print("Transformed Data Shape: {}".format(features.shape))  
print("Model's Component Shape: {}".format(model.components_.shape))
```

```python
for component in model.components_:  
	component = component.reshape(28, 28)  
	show_as_image(component)
```
